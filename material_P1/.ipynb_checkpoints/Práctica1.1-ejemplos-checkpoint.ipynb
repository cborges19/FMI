## 1. Muestreo

El muestreo es el proceso de convertir una señal continua en el tiempo en una señal discreta, tomando muestras de la señal original en intervalos regulares.

### Ejemplo de Muestreo

Tomemos una señal sinusoidal $x(t) = \sin(2\pi f t)$, con una frecuencia de $f = 15 Hz$, y muestremos la señal con diferentes frecuencias de muestreo $F_s$.

```python
f = 15  # Frecuencia de la señal en Hz
t_cont1 = np.arange(0, 1 + 1/10, 1/10)
t_cont2 = np.arange(0, 1 + 1/100, 1/100)
t_cont3 = np.arange(0, 1 + 1/1000, 1/1000)
x_cont1 = np.sin(2 * np.pi * f * t_cont1)  # Señal continua
x_cont2 = np.sin(2 * np.pi * f * t_cont2)  # Señal continua
x_cont3 = np.sin(2 * np.pi * f * t_cont3)  # Señal continua


plt.figure()
plt.plot(t_cont1, x_cont1, label='S1')
plt.legend()
plt.figure()
plt.plot(t_cont2, x_cont3, label='S2')
plt.legend()
plt.figure()
plt.plot(t_cont3, x_cont1, label='S3')
plt.legend()
```

### Código:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
f = 15  # Frecuencia de la señal en Hz

# Muestreo a diferentes frecuencias
F = 100 
Fs1 = 100  # Frecuencia de muestreo (por encima de la frecuencia de Nyquist)
Fs2 = 30  # Frecuencia de muestreo (igual a la frecuencia de Nyquist)
Fs3 = 20   # Frecuencia de muestreo (por debajo de la frecuencia de Nyquist)
Tt1 = 1/(F*Fs1)
Tt2 = 1/(F*Fs2)
Tt3 = 1/(F*Fs3)


t_cont1 = np.arange(0, 1 + Tt1, Tt1)
x_cont1 = np.sin(2 * np.pi * f * t_cont1)  # Señal continua

t_cont2 = np.arange(0, 1 + Tt2, Tt2)
x_cont2 = np.sin(2 * np.pi * f * t_cont2)  # Señal continua

t_cont3 = np.arange(0, 1 + Tt3, Tt3)
x_cont3 = np.sin(2 * np.pi * f * t_cont3)  # Señal continua

# Función de muestreo
def sample_signal(t, x, Fs, Ft):
    
    # Generar los puntos de muestreo
    t_sampled = np.arange(0, t[-1] + 1/Fs, 1/Fs)  # Tiempo muestreado
    xs = x[::int(Ft/Fs)]  # Selección de valores muestreados
    ts = t_sampled[0:len(xs)]

    
    return ts, xs


# Muestreo a diferentes frecuencias
t_s1, x_s1 = sample_signal(t_cont1, x_cont1, Fs1, F*Fs1)
t_s2, x_s2 = sample_signal(t_cont2, x_cont2, Fs2, F*Fs2)
t_s3, x_s3 = sample_signal(t_cont3, x_cont3, Fs3, F*Fs3)

# Visualización
plt.figure(figsize=(12, 8))

# Señal continua
plt.subplot(3, 1, 1)
plt.plot(t_cont1, x_cont1, label='Señal continua')
plt.stem(t_s1, x_s1, linefmt='r-', markerfmt='ro', basefmt='r-', label=f'Muestreo a {Fs1} Hz')
plt.title('Muestreo por encima de la frecuencia de Nyquist')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t_cont2, x_cont2, label='Señal continua')
plt.stem(t_s2, x_s2, linefmt='g-', markerfmt='go', basefmt='g-', label=f'Muestreo a {Fs2} Hz')
plt.title('Muestreo a la frecuencia de Nyquist')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t_cont3, x_cont3, label='Señal continua')
plt.stem(t_s3, x_s3, linefmt='b-', markerfmt='bo', basefmt='b-', label=f'Muestreo a {Fs3} Hz')
plt.title('Muestreo por debajo de la frecuencia de Nyquist')
plt.legend()

plt.tight_layout()
plt.show()
```

### Explicación:
- La primera señal muestra un muestreo a una frecuencia que está por encima de la frecuencia de Nyquist.
- La segunda muestra un muestreo exacto a la frecuencia de Nyquist.
- La tercera muestra un ejemplo de solapamiento cuando el muestreo es insuficiente.

## 2. Cuantización

La cuantización es el proceso de convertir una señal muestreada (que aún tiene valores continuos) en una señal con valores discretos en un conjunto finito de niveles.

### Implementación Manual del Cuantizador

A continuación implementaremos un cuantizador sin funciones predefinidas, para una señal muestreada.

### Código:

```python
# Parámetros del cuantizador
num_levels = 16  # Número de niveles de cuantización
xmin, xmax = -1, 1  # Rango de la señal
delta = (xmax - xmin) / num_levels  # Paso de cuantización

# Función de cuantización manual
def cuantizador(x, xmin, xmax, num_levels):
    x_q = np.clip(x, xmin, xmax)  # Limitar los valores al rango de la señal
    x_q = np.round((x_q - xmin) / delta) * delta + xmin  # Cuantización
    return x_q

# Cuantización de la señal muestreada
xq_s1 = cuantizador(x_s1, xmin, xmax, num_levels)

# Visualización
plt.figure(figsize=(10, 6))
plt.stem(t_s1, xq_s1, linefmt='r-', markerfmt='ro', basefmt='r-', label='Señal cuantizada')
plt.stem(t_s1, x_s1, linefmt='b-', markerfmt='bo', basefmt='b-', label='Señal muestreada')
plt.title('Cuantización de la señal muestreada')
plt.legend()
plt.grid()
plt.show()

# Cálculo del error de cuantización (MSE)
error_cuantizacion = np.mean((x_s1 - xq_s1)**2)

print(f"Error de cuantización (MSE): {error_cuantizacion:.6f}")
```

### Explicación:
El proceso de cuantización asigna cada muestra muestreada a un nivel de cuantización en un rango discreto, lo que introduce un error conocido como ruido de cuantización.

## 3. Reconstrucción

La reconstrucción es el proceso de convertir una señal cuantizada y muestreada de nuevo en una señal continua. El interpolador más común es el filtro paso bajo ideal, pero aquí usaremos una interpolación lineal simple para demostrar el concepto.

### Código:

```python
from scipy.interpolate import interp1d

# Función sinc para la interpolación
def sinc_interpolacion(t, ts, xs):
    """
    Interpolación usando la función sinc
    t: puntos donde evaluar la interpolación (tiempo continuo)
    ts: puntos de muestreo (tiempo discreto)
    xs: valores muestreados (señal discreta)
    """
    # Inicializar la señal reconstruida
    x_reconstruido = np.zeros_like(t)
    
    # Interpolar usando la suma ponderada de las funciones sinc
    for i in range(len(xs)):
        x_reconstruido += xs[i] * np.sinc((t - ts[i]) / (ts[1] - ts[0]))
    
    return x_reconstruido

# Interpolación lineal para la reconstrucción
lin_rec = interp1d(t_s1, xq_s1, kind='linear')
cub_rec =  interp1d(t_s1, xq_s1, kind='cubic')
near_rec =  interp1d(t_s1, xq_s1, kind='nearest')
# Tiempo continuo para reconstruir
x_lin = lin_rec(t_cont1)
x_cub = cub_rec(t_cont1)
x_near = near_rec(t_cont1)

x_sinc = sinc_interpolacion(t_cont1, t_s1, xq_s1)

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(t_cont1, x_cont1, label='Señal continua original')
plt.plot(t_cont1, x_lin, 'r--', label='Señal reconstruida')
#plt.stem(t_s1, xq_s1, linefmt='g-', markerfmt='go', basefmt='g-', label='Señal cuantizada')
plt.title('Reconstrucción de la señal cuantizada')
plt.legend()
plt.show()


# Cálculo del error de reconstrucción

err_lineal = np.mean((x_lin - x_cont1)**2)
err_cub = np.mean((x_cub - x_cont1)**2)
err_near = np.mean((x_near - x_cont1)**2)
err_sinc = np.mean((x_sinc - x_cont1)**2)
print(f"Error de reconstrucción lineal: {err_lineal:.6f}")
print(f"Error de reconstrucción cúbica: {err_cub:.6f}")
print(f"Error de reconstrucción: el valor más cercano: {err_near:.6f}")
print(f"Error de reconstrucción sinc: {err_sinc:.6f}")

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(t_cont1, x_cont1, label='Señal continua original')
plt.plot(t_cont1, x_sinc, 'r--', label='Señal reconstruida')
#plt.stem(t_s1, xq_s1, linefmt='g-', markerfmt='go', basefmt='g-', label='Señal cuantizada')
plt.title('Reconstrucción de la señal cuantizada')
plt.legend()
plt.show()
```

### Explicación:
La interpolación lineal permite reconstruir una señal continua a partir de muestras discretas cuantizadas. Sin embargo, el posfiltro ideal (que elimina componentes de alta frecuencia) es necesario para una reconstrucción perfecta.

## 4. Solapamiento

El solapamiento ocurre cuando una señal se muestrea por debajo de la frecuencia de Nyquist, lo que provoca que las altas frecuencias se plieguen y se mezclen con las bajas frecuencias.

### Ejemplo de solapamiento:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros para el solapamiento
f = 10  # Frecuencia de la señal en Hz

# Muestreo a diferentes frecuencias
F = 100 
Fas = 15  # Frecuencia de muestreo 
Tat = 1/(F*Fas)


tas = np.arange(0, 1 + Tat, Tat)
xas = np.sin(2 * np.pi * f * tas)  # Señal continua

t_alias, x_alias = sample_signal(tas, xas, Fas, F*Fas)

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(tas, xas, label=f'Señal continua de {f} Hz')
plt.stem(t_alias, x_alias, linefmt='r-', markerfmt='ro', basefmt='r-', label=f'Muestreo a {Fas} Hz')
plt.title('Solapamiento: Muestreo por debajo de la frecuencia de Nyquist')
plt.legend()
plt.show()


x_sinc = sinc_interpolacion(tas, t_alias, x_alias)

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(tas, xas, label='Señal continua original')
plt.plot(tas, x_sinc, 'r--', label='Señal reconstruida')
plt.title('Reconstrucción de la señal cuantizada')
plt.legend()
plt.show()

err_sinc = np.mean((x_sinc - xas)**2)
print(f"Error de reconstrucción sinc: {err_sinc:.6f}")
```

### Explicación:
En este ejemplo, la señal de 10 Hz es muestreada a 15 Hz (por debajo de la frecuencia de Nyquist), lo que genera solapamiento. La señal muestreada parece tener una frecuencia más baja de lo que realmente tiene.
