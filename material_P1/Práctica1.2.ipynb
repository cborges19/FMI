# Análisis de Distribuciones Aleatorias


### 1. ¿Cómo varían las medias y varianzas entre las distribuciones?

* **Media ($\mu$):** * En la **Distribución Normal**, la media se localiza en el centro de la campana (en tus gráficas es $0$). Es el valor más probable.
    * En la **Distribución Uniforme $[0, 1]$**, la media se sitúa en el punto medio del rango ($0.5$).
    * En la **Distribución de Enteros**, la media es el promedio de los valores posibles (aproximadamente $4.5$ si el rango es $0-9$).
* **Varianza ($\sigma^2$):** * En la **Normal**, la varianza determina qué tan "ancha" es la campana; los datos están muy concentrados cerca de la media.
    * En la **Uniforme**, la varianza es mayor en relación al rango, ya que los datos no tienen preferencia por el centro y están "esparcidos" uniformemente hasta los límites.

---

### 2. ¿Qué diferencias visuales existen entre la distribución uniforme y la normal?

| Característica | Distribución Uniforme | Distribución Normal (Gaussiana) |
| :--- | :--- | :--- |
| **Histograma** | Forma rectangular. Todas las barras tienen aproximadamente la misma altura. | Forma de campana. Las barras son altas al centro y bajan hacia los extremos. |
| **Gráfico de Líneas** | Los puntos están uniformemente distribuidos entre $0$ y $1$ sin una zona de mayor densidad. | Se observa una mayor densidad de líneas alrededor del valor $0$, con picos aislados hacia arriba y abajo. |
| **Límites** | Tiene límites definidos y constantes (ej. no hay valores menores a $0$ o mayores a $1$). | No tiene límites estrictos, pero la probabilidad de encontrar valores más allá de $\pm 3$ desviaciones estándar es mínima. |



---

### 3. ¿Qué se observa en el histograma al aumentar el número de muestras generadas?

Al aumentar el número de muestras (por ejemplo, pasar de $100$ a $10,000$ datos), se observan los siguientes cambios:

1.  **Reducción del "ruido":** Las fluctuaciones aleatorias entre barras adyacentes disminuyen. El perfil del histograma se vuelve más suave.
2.  **Convergencia Teórica:** El histograma comienza a parecerse cada vez más a la línea negra continua (la función de densidad teórica). 
    * En la uniforme, la parte superior se vuelve una línea horizontal plana.
    * En la normal, la curva se vuelve perfectamente simétrica y definida.
3.  **Estabilidad:** Los estimadores estadísticos (media y varianza calculadas) se acercan cada vez más a los valores teóricos ideales (Ley de los Grandes Números).

### 1. ¿Es el proceso ergódico? Justifica tu respuesta.

**No, el proceso no es ergódico.**

Un proceso es ergódico si sus promedios temporales coinciden con sus promedios estadísticos (de conjunto). Al observar las gráficas:
* **Media Temporal:** Se observa que la media a lo largo del tiempo para cada realización fluctúa cerca de **0**. Esto es lógico, ya que el promedio de una señal senoidal pura en varios ciclos tiende a cero.
* **Media Estadística:** El promedio calculado a través de las diferentes realizaciones mantiene la forma de la **señal senoidal** original.
* **Conclusión:** Como la media temporal (un valor constante cercano a 0) es distinta a la media estadística (una función senoidal del tiempo), el proceso **no es ergódico**.

---

### 2. ¿Podemos decir que el proceso es estacionario en el sentido amplio (WSS)? ¿Por qué?

**No, el proceso no es estacionario en el sentido amplio (WSS).**

Para que un proceso sea WSS (Wide-Sense Stationary), debe cumplir dos condiciones principales:
1.  **Media constante:** La media estadística $\mu(t)$ debe ser independiente del tiempo ($t$).
2.  **Autocorrelación dependiente solo de $\tau$:** La relación entre dos puntos solo debe depender de la distancia temporal entre ellos.

**Justificación basada en las imágenes:**
* La gráfica de **Media Estadística** muestra claramente que la media **varía en el tiempo**. 
* Dado que $\mu(t)$ no es constante (cambia según el punto del ciclo senoidal en el que nos encontremos), el proceso no puede ser estacionario.

---

### Resumen 
* Para mejorar la **Media Estadística**: Se debe aumentar el **número de realizaciones** (muestrear más señales independientes) para que el ruido promedio se anule y quede la señal pura.
* Para mejorar la **Media Temporal**: Se debe aumentar el **número de muestras en el tiempo** (duración de la realización), lo que permite que el promedio del ruido y de los ciclos senoidales converja mejor.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Parámetros del proceso
num_realizaciones = 50  # Número de realizaciones (para media estadística)
num_muestras = 1000  # Número de muestras en el tiempo (para media temporal)
t = np.linspace(0, 10, num_muestras)  # Vector de tiempo
amplitud = 5
frecuencia = 1
fase = 0
desviacion_estandar = 1
fs = num_muestras / (t[-1] - t[0])  # Frecuencia de muestreo

# Generamos un proceso estocástico: Ruido blanco
def generar_proceso_estocastico(num_realizaciones, num_muestras):
    # Relleno con ruido normal (media 0, sd=1) en una matriz de (R x M)
    return np.random.normal(0, desviacion_estandar, (num_realizaciones, num_muestras))

# Generamos un proceso estocástico: Señal sinusoidal más ruido
def generar_proceso_estocastico_seno(num_realizaciones, num_muestras, t, amplitud, frecuencia, fase, desviacion_estandar):
    # Genero la componente determinista para el vector de tiempo
    seno = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)
    # Genero el ruido aleatorio para todas las realizaciones
    ruido = np.random.normal(0, desviacion_estandar, (num_realizaciones, num_muestras))
    return seno + ruido

# Generamos un proceso estocástico: Ruido con cambio de media a mitad del tiempo
def generar_proceso_estocastico_media_variable(num_realizaciones, num_muestras, desviacion_estandar=1):
    """
    Genera un proceso estocástico donde la mitad del tiempo tiene media 0
    y la otra mitad tiene media 1.
    """
    mitad = num_muestras // 2
    # Primera mitad: ruido con media 0 para todas las realizaciones
    ruido_1 = np.random.normal(0, desviacion_estandar, (num_realizaciones, mitad))
    # Segunda mitad: ruido con media 1 para todas las realizaciones
    ruido_2 = np.random.normal(1, desviacion_estandar, (num_realizaciones, num_muestras - mitad))
    # Concatenamos las dos mitades por el eje del tiempo (axis=1)
    return np.concatenate((ruido_1, ruido_2), axis=1)


def generar_proceso_estocastico_coseno(num_realizaciones, num_muestras, t, amplitud=1, frecuencia=5):
    """
    Genera un proceso estocástico estacionario en sentido amplio: 
    un coseno con fase aleatoria uniforme.
    """
    X = np.zeros((num_realizaciones, num_muestras))
    for i in range(num_realizaciones):
        phi = np.random.uniform(0, 2 * np.pi)  # Fase aleatoria uniforme entre 0 y 2pi
        # Calculo el coseno para esta fase específica
        X[i, :] = amplitud * np.cos(2 * np.pi * frecuencia * t + phi)
    return X


# --- Selección de Proceso (puedes descomentar el que necesites) ---
X = generar_proceso_estocastico(num_realizaciones, num_muestras)
#X = generar_proceso_estocastico_seno(num_realizaciones, num_muestras, t, amplitud, frecuencia, fase, desviacion_estandar)
#X = generar_proceso_estocastico_media_variable(num_realizaciones, num_muestras)
#X = generar_proceso_estocastico_coseno(num_realizaciones, num_muestras, t, amplitud, frecuencia)

# 1. Media temporal: Promedio a lo largo del tiempo para cada fila (realización)
media_temporal = np.mean(X, axis=1)

# 2. Media estadística: Promedio a lo largo de las realizaciones para cada instante (columna)
media_estadistica = np.mean(X, axis=0)

# 3. Autocorrelación
def autocorrelacion(signal):
    # Calculo la correlación y normalizo por la longitud
    return np.correlate(signal, signal, mode='full') / len(signal)

autocorrelacion_primera_realizacion = autocorrelacion(X[0])
# Calculo la media de las autocorrelaciones de todas las filas
autocorrelacion_media = np.mean([autocorrelacion(X[i]) for i in range(num_realizaciones)], axis=0)
desplazamiento = np.arange(-num_muestras + 1, num_muestras)

# Gráficas de medias y autocorrelaciones
plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.plot(np.arange(num_realizaciones), media_temporal, color='red')
plt.title("Media Temporal (por realización)")
plt.xlabel("Realización")
plt.ylabel("Media Temporal")
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(t, media_estadistica, color='blue')
plt.title("Media Estadística (por instante de tiempo)")
plt.xlabel("Tiempo")
plt.ylabel("Media Estadística")
plt.grid(True)

plt.subplot(3, 2, 3)
plt.plot(desplazamiento, autocorrelacion_primera_realizacion, color='green')
plt.title("Autocorrelación de la primera realización")
plt.xlabel("Desplazamiento")
plt.ylabel("Autocorrelación")
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(desplazamiento, autocorrelacion_media, color='orange')
plt.title("Autocorrelación media de todas las realizaciones")
plt.xlabel("Desplazamiento")
plt.ylabel("Autocorrelación Media")
plt.grid(True)

plt.tight_layout()
plt.show()

# -----------------------------------------------------------
# 5. Densidad Espectral de Potencia (PSD)
# -----------------------------------------------------------

# Opción 1: PSD a partir de la autocorrelación media (definición teórica)
fft_autocorr = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(autocorrelacion_media)))
frecuencias_fft = np.fft.fftshift(np.fft.fftfreq(len(autocorrelacion_media), d=1/fs))
PSD_teorica = np.real(fft_autocorr)

# Opción 2: PSD con método de Welch (más estable)
f_welch, Pxx_welch = welch(X[0], fs=fs, nperseg=256)
# Media de las PSD de todas las realizaciones usando Welch
Pxx_welch_media = np.mean([welch(X[i], fs=fs, nperseg=256)[1] for i in range(num_realizaciones)], axis=0)

# Graficar la PSD
plt.figure(figsize=(10, 5))
plt.plot(frecuencias_fft, PSD_teorica, label='PSD teórica')
plt.plot(f_welch, Pxx_welch_media, label='PSD Welch media', color='orange')
plt.title('Densidad Espectral de Potencia (PSD)')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('PSD [V^2/Hz]')
plt.legend()
plt.grid(True)
plt.show()
```
